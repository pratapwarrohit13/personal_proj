```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
path = 'D://DS//Stater_Projects//housing.txt'
```

```python
df = pd.read_csv(path, header=None)
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df = df.values
```

```python
features, target = df[:,:-1], df[:,-1]
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, train_size=0.33,random_state=12345)
```

```python
scaler = MinMaxScaler()
scaler.fit(features_train)
```

```python
features_train = scaler.transform(features_train)
features_test = scaler.transform(features_test)
```

```python
print(features_train.shape, features_test.shape, target_train.shape, target_test.shape)
```

```python
n_features = features_train.shape[1]
```

```python
model = Sequential()
model.add(Dense(20, kernel_initializer='he_normal', activation='relu', input_dim=n_features))
model.add(Dense(5, kernel_initializer='he_normal', activation='relu'))
model.add(Dense(1))
```

```python
opt = Adam(learning_rate=0.01, beta_1=0.85, beta_2=0.999)
model.compile(optimizer=opt, loss='mse')
```

```python
model.fit(features_train, target_train, verbose=2, epochs=300, batch_size=16)
```

```python
prediction = model.predict(features_test, verbose=0)
```

```python
score = mean_absolute_error(target_test, prediction)
print('MAE: %.3f' % score)
```
