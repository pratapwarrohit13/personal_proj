```python
import pandas as pd
from sklearn.model_selection import train_test_split
from autokeras import StructuredDataRegressor
from numpy import asarray
```

```python
path = 'D://DS//Stater_Projects//car_insurance_payout.txt'
```

```python
df = pd.read_csv(path,header=None)
```

```python
df.head()
```

```python
df = df.values
```

```python
df = df.astype('float32')
```

```python
features, target = df[:, :-1], df[:, -1]
print(features.shape, target.shape)
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.33, random_state=12345)
print(features_train.shape, features_test.shape, target_train.shape, target_test.shape)
```

```python
autoks = StructuredDataRegressor(max_trials=15, loss='mean_absolute_error')
```

```python
autoks.fit(x=features_train, y=target_train, verbose=0)
```

```python
mae, _ = autoks.evaluate(features_test, target_test, verbose=0)
print('MAE: %.3f' % mae)
```

```python
feature_new = asarray([[108]]).astype('float32')
pred = autoks.predict(feature_new)
```

```python
print('Predicted: %.3f' % pred[0])
```

```python
model = autoks.export_model()
```

```python
model.summary()
```

```python
model.save('model_insurance.h5')
```
